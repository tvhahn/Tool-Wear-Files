```python
import numpy as np
import pandas as pd
import os
from pathlib import Path
```

```python
folder_path = Path("sample_csv_combine")

i = 0
for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        if i == 0:
            file_path = folder_path / file           
            df = pd.read_csv(file_path)
            i += 1
        else:
            file_path = folder_path / file
            df = df.append(pd.read_csv(file_path))
        print(file)
```
